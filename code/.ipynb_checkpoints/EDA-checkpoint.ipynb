```python
#imports
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import pickle
pd.set_option('display.max_rows', 500)
```

```python
# Read the csv data files from the instacart-market-basket-analysis directory
orders = pd.read_csv('../instacart-market-basket-analysis/orders.csv')
products = pd.read_csv('../instacart-market-basket-analysis/products.csv')
prior_ = pd.read_csv('../instacart-market-basket-analysis/order_products__prior.csv')
train = pd.read_csv('../instacart-market-basket-analysis/order_products__train.csv')
aisles = pd.read_csv('../instacart-market-basket-analysis/aisles.csv')
departments = pd.read_csv('../instacart-market-basket-analysis/departments.csv')
```

```python
# Read the pickle file
with open('../data/instacart_data.pkl', 'rb') as pickle_in:
    df = pickle.load(pickle_in)
```

```python
# Replace NaN with 0 in the final aggregate data file
df = df.replace(np.nan, 0)
```

```python
train_dow = train.merge(orders, on='order_id')

plt.title('Day of week customers Reordered')
C= train_dow[['reordered','order_dow']].groupby('order_dow').mean()
sns.barplot(x=C.index[0:], y=C['reordered'], color='orange')
# plt.ylim(0,1)
plt.show()
```

```python
#How many prior orders are available based on order_numer?
plt.figure(figsize=(27,5))
sns.countplot(x='order_number', data = orders.loc[orders.eval_set=='prior'], palette='coolwarm')
plt.show()
```

```python
# Effect of hour of day on order volume
sns.countplot(x='order_hour_of_day', data =orders)
plt.title("Purchase Hour Distribution")
plt.xlabel("Hour of the Day")
plt.show()
```

```python
# Effect of day of the w on order volume
sns.countplot(x='order_dow', data =orders)
plt.show()
```

```python
#List of best selling products
train_product = train.merge(products, on='product_id')

top_products = train_product.product_name.value_counts()[0:15]
bottom_products = train_product.product_name.value_counts()[-15:]
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x= top_products.index, y=top_products )
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x= bottom_products.index[0:25], y=bottom_products )
plt.xticks(rotation=90)
plt.show()
```

```python
# frequency of orders
order_frequency=orders.groupby("order_id")["days_since_prior_order"].max().reset_index()
order_frequency.head()

plt.figure(figsize=(20,15))
sns.countplot(order_frequency["days_since_prior_order"])
plt.title("How often Customers Purchase")
plt.xlabel("Day Since Prior Order")
plt.ylabel("Count")
```
