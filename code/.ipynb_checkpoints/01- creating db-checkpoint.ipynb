# Aggregate for all customers orders

```python
# STEP 1: Get the days count
# count of number of times a customer buys an item
c.execute(""" 
create table cust_orders as

with num_ordered as
(select o.user_id, p.product_id, count(p.product_id) as times_ordered 
from orders o
    inner join order_products__prior p on o.order_id = p.order_id
group by user_id, product_id),
total_orders as
(select user_id, max(order_number) as total_orders 
from orders o
where eval_set = 'prior'
group by user_id),
avg_days as
(select user_id, min(days_since_prior_order) min_days_bet, max(days_since_prior_order) max_days_bet, avg(days_since_prior_order) avg_days_bet 
from orders
where order_number > 1
group by user_id)
Select o.user_id, o.product_id, o.times_ordered, t.total_orders, d.min_days_bet, d.max_days_bet, d.avg_days_bet
from num_ordered o 
    inner join total_orders t on o.user_id = t.user_id
    inner join avg_days d on d.user_id = t.user_id
""")
```

```python
# STEP 2: Get days between each order item - interim temp table
c.execute("""
create table items_order_by_user as

with temp_order as
(
select order_id, user_id, order_number, order_number, days_since_prior_order,
         SUM(days_since_prior_order) OVER
         (PARTITION BY user_id ORDER BY user_id, order_number) AS running_total_days
from orders 
)
select o.order_id, o.user_id, o.days_since_prior_order, p.product_id, o.running_total_days, o.order_number, 
         o.running_total_days -LAG(o.running_total_days, 1) OVER
         (PARTITION BY o.user_id, p.product_id ORDER BY o.user_id, p.product_id, o.order_number) AS difference
from temp_order o
    inner join order_products__prior p on o.order_id = p.order_id
order by o.user_id, p.product_id, o.order_number
""")
```

```python
# STEP 3: Get the final differences between a customer's order of an item
c.execute("""
create table items_reorder_days as

select user_id, product_id, min(difference) min_reorder_days, 
max(difference) max_reorder_days, avg(difference) avg_reorder_days
from items_order_by_user 
group by user_id, product_id
""")
```

```python
# STEP 4: Add all above columns together
c.execute("""
create table all_orders as

select c.user_id, c.product_id, c.times_ordered, c.total_orders, 
        CAST(c.times_ordered AS float) / CAST(c.total_orders AS float) pct_reorder,
        c.min_days_bet, c.max_days_bet, c.avg_days_bet,
        d.min_reorder_days, d.max_reorder_days, d.avg_reorder_days
from cust_orders c
    inner join items_reorder_days d on c.user_id = d.user_id and c.product_id = d.product_id
""")
```

```python
# STEP 5: Get Train data and add to above dataset
# Order ID will not match, therefore we need to join on user and product
c.execute("""
create table train as

select o.user_id, t.product_id, t.reordered 
from orders o
    inner join order_products__train t on o.order_id = t.order_id 
order by o.user_id, t.product_id
""")

```

```python
# STEP 6: Merge train and all orders datasets to give final dataset
c.execute("""
create table insta_final as

select o.*, CASE WHEN t.reordered IS NULL THEN 0 ELSE t.reordered END as target
from all_orders o
    left join train t on o.user_id = t.user_id and o.product_id = t.product_id
""")
```

# Modify above to also get the values for the last 5 orders

```python
# STEP 1: reorder the order_number to get the last 5 orders 
c.execute("""
create table last_5orders as

with temp_orders as 
(select order_id, user_id, days_since_prior_order, row_number() over(partition by user_id order by order_number desc) new_num
, order_number
from orders o
where eval_set = 'prior')
select order_id, user_id, days_since_prior_order, 
    row_number() over(partition by user_id order by new_num desc) new_order_num
from temp_orders
where new_num <= 5
""")
```

```python
# STEP 2: Get the days count
# count of number of times a customer buys an item
c.execute(""" 
create table cust_last5_orders as

with num_ordered as
(select o.user_id, p.product_id, count(p.product_id) as times_ordered 
from last_5orders o
    inner join order_products__prior p on o.order_id = p.order_id
group by user_id, product_id),
total_orders as
(select user_id, max(new_order_num) as total_orders 
from last_5orders o
group by user_id),
avg_days as
(select user_id, min(days_since_prior_order) min_days_5odrs, max(days_since_prior_order) max_days_5odrs, avg(days_since_prior_order) avg_days_5odrs 
from last_5orders
where new_order_num > 1
group by user_id)
Select o.user_id, o.product_id, o.times_ordered, t.total_orders, d.min_days_5odrs, d.max_days_5odrs, d.avg_days_5odrs
from num_ordered o 
    inner join total_orders t on o.user_id = t.user_id
    inner join avg_days d on d.user_id = t.user_id
""")
```

```python
# STEP 3: Get days between each order item - interim temp table
c.execute("""
create table last5_order_by_user as

with temp_order as
(
select order_id, user_id, new_order_num, days_since_prior_order,
         SUM(days_since_prior_order) OVER
         (PARTITION BY user_id ORDER BY user_id, new_order_num) AS running_total_days
from last_5orders 
)
select o.order_id, o.user_id, o.days_since_prior_order, p.product_id, o.running_total_days, o.new_order_num, 
         o.running_total_days -LAG(o.running_total_days, 1) OVER
         (PARTITION BY o.user_id, p.product_id ORDER BY o.user_id, p.product_id, o.new_order_num) AS difference
from temp_order o
    inner join order_products__prior p on o.order_id = p.order_id
order by o.user_id, p.product_id, o.new_order_num
""")
```

```python
# STEP 4: Get the final differences between a customer's order of an item
c.execute("""
create table items_reorder_5last as

select user_id, product_id, min(difference) min_reorder_5days, 
max(difference) max_reorder_5days, avg(difference) avg_reorder_5days
from last5_order_by_user 
group by user_id, product_id
""")
```

```python
# STEP 5: Add all above columns together
c.execute("""
create table last5_orders as

select c.user_id, c.product_id, c.times_ordered, c.total_orders, 
        CAST(c.times_ordered AS float) / CAST(c.total_orders AS float) pct_reorder,
        c.min_days_5odrs, c.max_days_5odrs, c.avg_days_5odrs,
        d.min_reorder_days min_reorder_5days, d.max_reorder_days max_reorder_5days, d.avg_reorder_days avg_reorder_5days
from cust_last5_orders c
    inner join items_reorder_5last d on c.user_id = d.user_id and c.product_id = d.product_id
""")
```

```python
# STEP 6: Get Train data and add to above dataset
# Order ID will not match, therefore we need to join on user and product
# NOT NEEDED HERE
c.execute("""
create table last5_train as

select o.user_id, t.product_id, t.reordered 
from last_5orders o
    inner join order_products__train t on o.order_id = t.order_id 
order by o.user_id, t.product_id
""")
```

```python
# STEP 7: Merge train and all orders datasets to give final dataset
c.execute("""
create table instacart_data as

select f.*, o.times_ordered last5_times_item_pur, o.total_orders last5_total_orders, o.pct_reorder last5_pct_order
    , o.min_days_5odrs, o.max_days_5odrs, o.avg_days_5odrs, o.min_reorder_5days, o.max_reorder_5days, o.avg_reorder_5days
from insta_final f
    left join last5_orders o on f.user_id = o.user_id and f.product_id = o.product_id
order by f.user_id, f.product_id
""")
```

# Final output file

```python
# STEP 1:
# Calculate the average number of items a customer purchase in all their orders
c.execute("""
CREATE table avg_items as

select user_id, round(count(*) / max(order_number), 0) avg_items_ordered
from items_order_by_user
group by user_id
""")
```

```python
# STEP 2:
# Calculate the average number of items a customer purchase in their last 5 orders
c.execute("""
CREATE table last5_avg_items as

select user_id, round(count(*) / max(new_order_num), 0) last5_avg_items
from last5_order_by_user 
group by user_id
""")
```

```python
# STEP 3:
# Add Department, keeping the user_id and product_id for EDA 
c.execute("""
CREATE table instacart_final as

SELECT i.user_id || '-' || i.product_id as user_product, d.department, i.*, a.avg_items_ordered, l.last5_avg_items 
FROM instacart_data i
     JOIN products p ON i.product_id = p.product_id
     JOIN departments d ON p.department_id = d.department_id
     JOIN avg_items a ON i.user_id = a.user_id
     JOIN last5_avg_items l ON i.user_id = l.user_id
""")
```
